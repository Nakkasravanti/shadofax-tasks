```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Provide the correct path to the CSV file (use a raw string or double backslashes)
file_path = r'C:\Users\sravanthi\Downloads\delhiaqi.csv'

# Load the dataset
df = pd.read_csv(file_path)

# Display the first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values
df = df.dropna()

# Convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# List of key pollutants
pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']

# Check if all pollutants are in the DataFrame columns
for pollutant in pollutants:
    if pollutant not in df.columns:
        print(f"Warning: {pollutant} is not a column in the DataFrame")

# Plotting the pollutants
plt.figure(figsize=(12, 8))
for pollutant in pollutants:
    if pollutant in df.columns:
        sns.lineplot(x='date', y=pollutant, data=df, label=pollutant)

plt.title('Trends of Key Pollutants Over Time')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Provide the correct path to the CSV file (use a raw string or double backslashes)
file_path = r'C:\Users\sravanthi\Downloads\delhiaqi.csv'

# Load the dataset
df = pd.read_csv(file_path)

# Display the first few rows
print("First few rows of the dataframe:")
print(df.head())

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Fill or drop missing values
df = df.dropna()

# Check if 'date' column exists and convert it to datetime format
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
else:
    print("The 'date' column is not present in the dataset.")

# Verify the existence of the pollutant columns
pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']
missing_columns = [pollutant for pollutant in pollutants if pollutant not in df.columns]
if missing_columns:
    print(f"The following pollutant columns are missing in the dataframe: {missing_columns}")
else:
    # Extracting month and year
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
 # Grouping by month to see seasonal variations
    monthly_avg = df.groupby('month')[pollutants].mean()

    # Plotting seasonal variations
    plt.figure(figsize=(12, 8))
    monthly_avg.plot(kind='bar', stacked=True)
    plt.title('Seasonal Variations in AQI')
    plt.xlabel('Month')
    plt.ylabel('Average Concentration')
    plt.legend(title='Pollutants')
    plt.show()
```

```python
import pandas as pd
import folium

# Sample data for AQI monitoring stations
data = {
    'latitude': [28.7041, 28.6692, 28.5272],
    'longitude': [77.1025, 77.1235, 77.2090]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Creating a map of Delhi
delhi_map = folium.Map(location=[28.6139, 77.2090], zoom_start=10)

# Adding markers for different AQI monitoring stations
for index, row in df.iterrows():
    folium.CircleMarker(
        [row['latitude'], row['longitude']],
        radius=5,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(delhi_map)

# Saving the map to an HTML file
delhi_map.save('delhi_aqi_map.html')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for illustration
data = {
    'date': pd.date_range(start='2022-10-01', end='2022-11-30', freq='D'),
    'PM2.5': [50 + i for i in range(61)]  # Example PM2.5 data
}

# Creating DataFrame
df = pd.DataFrame(data)

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the month from the 'date' column
df['month'] = df['date'].dt.month

# Filter data around Diwali period (e.g., October and November)
diwali_data = df[(df['month'] == 10) | (df['month'] == 11)]

# Plotting the AQI around Diwali
plt.figure(figsize=(12, 8))
sns.lineplot(x='date', y='PM2.5', data=diwali_data)
plt.title('AQI Around Diwali')
plt.xlabel('Date')
plt.ylabel('PM2.5 Concentration')
plt.show()
```
